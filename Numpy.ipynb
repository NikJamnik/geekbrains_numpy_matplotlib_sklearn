###  Task 1

```
import numpy as np
```

```
a = np.array([[1, 6],
              [2, 8],
              [3, 11],
              [3, 10],
              [1, 7]], dtype = np.float32)
```

```
mean_a = a.mean(axis = 0)
mean_a
```

### Task 2

```
a_centered = a - mean_a
a_centered
```

### Task 3

```
a_centered_sp = np.dot(a_centered[:, 0], a_centered[:, 1])
a_centered_sp /= (a_centered.shape[0] - 1)
a_centered_sp
```

### Task 4

```
cov_matrix = np.cov(m=a.T)
cov = cov_matrix[0, 1]
cov
```

### Task 1

```
import pandas as pd
```

```
authors = pd.DataFrame({'author_id': [1, 2, 3], 
                        'author_name': ['Тургенев', 'Чехов', 'Островский']})
authors
```

```
books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3],
                     'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 
                                    'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 
                                    'Таланты и поклонники'],
                     'price': [450, 300, 350, 500, 450, 370, 290]})
books
```

### Task 2

```
authors_price = pd.merge(authors, books, on='author_id', how='inner')
authors_price
```

### Task 3

```
top5 = authors_price.sort_values(by='price', ascending='False').head()
top5
```

### Task 4

```
authors_stat = pd.DataFrame({'author_name': ['Тургенев', 'Чехов', 'Островский']})
authors_stat['min_price'] = [authors_price.loc[authors_price.author_name == name, 'price'].min() for name in authors_stat.loc[:, 'author_name']]
authors_stat['max_price'] = [authors_price.loc[authors_price.author_name == name, 'price'].max() for name in authors_stat.loc[:, 'author_name']]
authors_stat['mean_price'] = [authors_price.loc[authors_price.author_name == name, 'price'].mean() for name in authors_stat.loc[:, 'author_name']]
authors_stat
```

### Task 5

```
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
```

```
book_info = pd.pivot_table(data=authors_price.drop(['author_id'], axis=1), 
                           index=['author_name'], 
                           columns=['cover'],
                           aggfunc=['sum'],
                           fill_value=0)
book_info.columns = ['мягкая', 'твердая']
book_info
```

```
book_info.to_pickle("book_info.pkl")
```

### V1

```
v1_theft = data.loc[data['Class'] == 1, 'V1'] / data.shape[0]
v1_integrity = data.loc[data['Class'] == 0, 'V1'] / data.shape[0]
```

```
plt.rcParams['figure.figsize'] = 8, 6
plt.hist(v1_theft, bins=20, density=True, alpha=0.5, color='red', label='Class 1')
plt.hist(v1_integrity, bins=20, density=True, alpha=0.5, color='gray', label='Class 0')
plt.legend()
plt.ylabel('Class', fontdict={'family': FONT_STYLE, 'fontsize': 16});
```
